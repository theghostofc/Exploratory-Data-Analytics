```python
# No. of siblings of children of Illeterate mothers.
import pandas as pd
import matplotlib.pyplot as plt
size = 30
df = pd.read_csv("../Data/nas.csv")
# df.head()
df = df.loc[:, ['Siblings', 'Mother.edu']]
df['Siblings'] = df['Siblings'].str.replace('Single', '1')
df = df.dropna()
print("=======")
df.info()
print(df['Siblings'].unique())
print(df['Mother.edu'].unique())
print("=======")
df['Siblings'] = df['Siblings'].str.extract('(\d+)', expand=True).astype(int)

# Approach #1
print("Approach #1 - Groupby-mean")
print("==========================")
df_1 = df.groupby('Mother.edu')
print(df_1['Siblings'].mean())

# Approach #2
print("Approach #2 - Pivot Table ")
print("==========================")
df_1 = df.pivot_table(values = 'Siblings', index = 'Mother.edu')
print(df_1.head(20))

# df_1.head()
```

```python
# How the average science marks vary with the father’s education level and age.
import pandas as pd
import matplotlib.pyplot as plt
size = 30
df = pd.read_csv("../Data/nas.csv")
# df.head()
df = df.loc[:, ['Age', 'Father.edu', 'Science..']]
#df['Siblings'] = df['Siblings'].str.replace('Single', '1')
df = df.dropna()
print("=======")
df.info()
print(df['Age'].unique())
print(df['Father.edu'].unique())
print("=======")
df['Age'] = df['Age'].str.extract('(\d+)', expand=True).astype(int)
df.head()
df = df[df['Age'] > 11]
# Approach #1
print("Approach #1 - Groupby-mean")
print("==========================")
df_1 = df.groupby(['Father.edu', 'Age'])
print(df_1['Science..'].mean())

# Approach #2
print("Approach #2 - Pivot Table ")
print("==========================")
df_1 = df.pivot_table(values = 'Science..', index = ['Father.edu', 'Age'])
print(df_1.head(20))

# # df_1.head()
# corr = df_1.corr()
# fig, ax = plt.subplots(figsize=(size, size))
# ax.matshow(corr)
# #plt.tight_layout()
# plt.xticks(range(len(corr.columns)), corr.columns, rotation='vertical');
# plt.yticks(range(len(corr.columns)), corr.columns);
```

```python
# ODI Hundereds count
import pandas as pd
df = pd.read_csv("../Data/odi-batting.csv")
df = df.loc[:, ['Player', 'Runs']]
df = df.dropna()
df = df[df['Runs'] >= 100]
df['Century'] = df['Runs'] >= 100
df_1 = df.pivot_table(values = 'Century', index = 'Player', aggfunc='count')
df_1.sort_values(by = 'Century', ascending = False, inplace = True)
print(df_1.head(20))
```

```python
# Fastest 100 - strike rate
import pandas as pd
df = pd.read_csv("../Data/odi-batting.csv")
df = df.loc[:, ['Player', 'Runs', 'Balls']]
df = df.dropna()
df = df[df['Runs'] >= 100]
df['StrikeRate'] = df['Runs'] * 100 / df['Balls']
df.sort_values(by = 'StrikeRate', ascending = False, inplace = True)
print(df.head(20))
```

```python
# In which year were the maximum number of centuries scored by Indian players?
import pandas as pd
df = pd.read_csv("../Data/odi-batting.csv")
df = df.loc[:, ['Country', 'Player', 'Runs', 'MatchDate']]
df = df.dropna()
df = df[(df['Runs'] >= 100) & (df['Country'] == 'India')]
df['Year'] = df['MatchDate'].str[-4:]
df['Century'] = df['Runs'] >= 100
df_1 = df.pivot_table(values = 'Century', index = 'Year', aggfunc='count')
df_1.sort_values(by = 'Century', ascending = False, inplace = True)
print(df_1.head(20))
```
